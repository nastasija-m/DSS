```python
# Read users_and_Top5artists.csv
users_and_Top5artists = pd.read_csv(r'C:\Users\nmaksimovi001\Downloads\DSS zipped\DSS\users_and_Top5artists.csv')
users_and_Top5artists
```

```python
users_and_Top5artists['artist_top_tracks'] = np.nan

for index, row in users_and_Top5artists.iterrows():
    artist_top_tracks = sp.artist_top_tracks(row['artist_id'])
    track_ids = []
    for track in artist_top_tracks['tracks']:
        track_ids.append(track['id'])
    
    users_and_Top5artists['artist_top_tracks'][index] = track_ids

# Explode the 'artist_top_tracks' column to create separate rows for each track
users_and_Top5artists = users_and_Top5artists.explode('artist_top_tracks')

# Reset index after exploding
users_and_Top5artists.reset_index(drop=True, inplace=True)
users_and_Top5artists
```

```python
# Get the audio features and add them to the dataframe
users_and_Top5artists['danceability'] = np.nan
users_and_Top5artists['energy'] = np.nan
users_and_Top5artists['key'] = np.nan
users_and_Top5artists['loudness'] = np.nan
users_and_Top5artists['mode'] = np.nan
users_and_Top5artists['speechiness'] = np.nan
users_and_Top5artists['acousticness'] = np.nan
users_and_Top5artists['instrumentalness'] = np.nan
users_and_Top5artists['liveness'] = np.nan
users_and_Top5artists['valence'] = np.nan
users_and_Top5artists['tempo'] = np.nan
users_and_Top5artists['duration_ms'] = np.nan
users_and_Top5artists['time_signature'] = np.nan

for index, row in users_and_Top5artists.iterrows():
    track_audio_features = sp.audio_features(row['artist_top_tracks'])
    for track in track_audio_features:
        users_and_Top5artists.at[index, 'danceability'] = track['danceability']
        users_and_Top5artists.at[index, 'energy'] = track['energy']
        users_and_Top5artists.at[index, 'key'] = track['key']
        users_and_Top5artists.at[index, 'loudness'] = track['loudness']
        users_and_Top5artists.at[index, 'mode'] = track['mode']
        users_and_Top5artists.at[index, 'speechiness'] = track['speechiness']
        users_and_Top5artists.at[index, 'acousticness'] = track['acousticness']
        users_and_Top5artists.at[index, 'instrumentalness'] = track['instrumentalness']
        users_and_Top5artists.at[index, 'liveness'] = track['liveness']
        users_and_Top5artists.at[index, 'valence'] = track['valence']
        users_and_Top5artists.at[index, 'tempo'] = track['tempo']
        users_and_Top5artists.at[index, 'duration_ms'] = track['duration_ms']
        users_and_Top5artists.at[index, 'time_signature'] = track['time_signature']

users_and_Top5artists
```
