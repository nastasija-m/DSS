```python
import numpy as np

def get_feature_range_quartiles(feature_values):
    
    lower_bound = np.percentile(feature_values, 25)
    upper_bound = np.percentile(feature_values, 75)

    # Calculate the interquartile range (ibr)
    ibr = upper_bound - lower_bound

    # Define the range excluding the first and last 25%
    min_value = lower_bound - 0.5 * ibr  # Adjust as needed
    max_value = upper_bound + 0.5 * ibr  # Adjust as needed

    # Clip the values to ensure they are within the range of the data
    min_value = max(min_value, np.min(feature_values))
    max_value = min(max_value, np.max(feature_values))

    return round(min_value,2), round(max_value,2)

feature = 'liveness'
min_key, max_key  = get_feature_range_quartiles(user0_tracks_all_tracks[feature])
print(min_key, max_key)

print(user0_tracks_all_tracks[(user0_tracks_all_tracks[feature]>=min_key) & (user0_tracks_all_tracks[feature]<=max_key)].count()/len(user0_tracks_all_tracks))

```

```python
import numpy as np
from scipy.stats import mode

def get_mode_interval(data, epsilon_percent):
    # Calculate mode
    mode_result = mode(data)
    mode_value = mode_result.mode[0]

    # Calculate epsilon value
    data_range = np.max(data) - np.min(data)
    epsilon = epsilon_percent * data_range

    # Define the interval
    lower_bound = mode_value - epsilon
    upper_bound = mode_value + epsilon

    return lower_bound, upper_bound, mode_value

# Example usage
your_data = user0_tracks_all_tracks['liveness']
epsilon_percent = 0.15  # You can adjust this percentage
lower, upper, mode_value = get_mode_interval(your_data, epsilon_percent)

print("Mode:", mode_value)
print(f"Interval around the mode: ({lower}, {upper})")
```

```python
import numpy as np
from scipy.stats import mode

def get_mode_range(data, percentage):
    # Calculate mode
    mode_result = mode(data)
    mode_value = mode_result.mode[0]

    # Calculate the lower and upper bounds around the mode
    lower_bound = mode_value - (percentage / 200) * np.ptp(data)
    upper_bound = mode_value + (percentage / 200) * np.ptp(data)

    return round(lower_bound, 2), round(upper_bound, 2)

# Example usage
your_data = user0_tracks_all_tracks['tempo']
percentage = 45
lower, upper = get_mode_range(your_data, percentage)

print("Mode:", mode_value)
print(f"Range around the mode (containing {percentage}% of the data): ({lower}, {upper})")

#print the density in the range for acousticness
print(user0_tracks_all_tracks[(user0_tracks_all_tracks['tempo']>=lower) & (user0_tracks_all_tracks['tempo']<=upper)].count()/len(user0_tracks_all_tracks))
```
