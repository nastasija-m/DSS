```python
# Your existing function
def get_feature_range_quartiles(feature_values):
    lower_bound = np.percentile(feature_values, 25)
    upper_bound = np.percentile(feature_values, 75)
    ibr = upper_bound - lower_bound
    min_value = lower_bound - 0.5 * ibr
    max_value = upper_bound + 0.5 * ibr
    min_value = max(min_value, np.min(feature_values))
    max_value = min(max_value, np.max(feature_values))
    return round(min_value, 2), round(max_value, 2)

# Create a new dataframe to store the results
result_df = pd.DataFrame(columns=['user_id'])

# Iterate through each feature and compute bounds
for feature in user_tracks.columns[2:]:  # Exclude 'user_id' from the iteration
    bounds = user_tracks.groupby('user_id')[feature].apply(get_feature_range_quartiles).apply(pd.Series)
    bounds.columns = [f'min_{feature}', f'max_{feature}']
    result_df = pd.concat([result_df, bounds], axis=1)

# Display the result
result_df.reset_index(inplace=True)

#drop the user_id column
result_df.drop(columns=['user_id'], inplace=True)

#rename the column index to user_id
result_df.rename(columns={'index':'user_id'}, inplace=True)
```
