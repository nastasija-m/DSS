```python
#read users_and_Top5artists.csv 
users_and_Top5artists = pd.read_csv(r'C:\Users\nasta\OneDrive - TU Eindhoven\Desktop\DSS\users_and_Top5artists.csv')
users_and_Top5artists
```

```python
#extract top 5 artist's top 5 tracks
users_and_Top5artists['artist_top_tracks'] = np.nan

for index, row in users_and_Top5artists.iterrows():
    artist_top_tracks = sp.artist_top_tracks(row['artist_id'])
    track_ids = []
    for track in artist_top_tracks['tracks']:
        track_ids.append(track['id'])
    
    users_and_Top5artists.at[index, 'artist_top_tracks'] = track_ids

# Explode the 'artist_top_tracks' column to create separate rows for each track
users_and_Top5artists = users_and_Top5artists.explode('artist_top_tracks')

# Reset index after exploding
users_and_Top5artists.reset_index(drop=True, inplace=True)

users_and_Top5artists
```

```python
#for each top track, get the audio features and add them to the dataframe
users_and_Top5artists['danceability'] = np.nan
users_and_Top5artists['energy'] = np.nan
users_and_Top5artists['key'] = np.nan
users_and_Top5artists['loudness'] = np.nan
users_and_Top5artists['mode'] = np.nan
users_and_Top5artists['speechiness'] = np.nan
users_and_Top5artists['acousticness'] = np.nan
users_and_Top5artists['instrumentalness'] = np.nan
users_and_Top5artists['liveness'] = np.nan
users_and_Top5artists['valence'] = np.nan
users_and_Top5artists['tempo'] = np.nan
users_and_Top5artists['duration_ms'] = np.nan
users_and_Top5artists['time_signature'] = np.nan

for index, row in users_and_Top5artists.iterrows():
    track_audio_features = sp.audio_features(row['artist_top_tracks'])
    for track in track_audio_features:
        users_and_Top5artists.at[index, 'danceability'] = track['danceability']
        users_and_Top5artists.at[index, 'energy'] = track['energy']
        users_and_Top5artists.at[index, 'key'] = track['key']
        users_and_Top5artists.at[index, 'loudness'] = track['loudness']
        users_and_Top5artists.at[index, 'mode'] = track['mode']
        users_and_Top5artists.at[index, 'speechiness'] = track['speechiness']
        users_and_Top5artists.at[index, 'acousticness'] = track['acousticness']
        users_and_Top5artists.at[index, 'instrumentalness'] = track['instrumentalness']
        users_and_Top5artists.at[index, 'liveness'] = track['liveness']
        users_and_Top5artists.at[index, 'valence'] = track['valence']
        users_and_Top5artists.at[index, 'tempo'] = track['tempo']
        users_and_Top5artists.at[index, 'duration_ms'] = track['duration_ms']
        users_and_Top5artists.at[index, 'time_signature'] = track['time_signature']

users_and_Top5artists
```

```python
#read user_tracks_high.csv and user_tracks_low.csv
users_and_Top5artists = pd.read_csv(r'C:\Users\nasta\OneDrive - TU Eindhoven\Desktop\DSS\users_top5artists_top5tracks.csv')
user_tracks_high = pd.read_csv(r'C:\Users\nasta\OneDrive - TU Eindhoven\Desktop\DSS\user_tracks_high.csv')
user_tracks_low = pd.read_csv(r'C:\Users\nasta\OneDrive - TU Eindhoven\Desktop\DSS\user_tracks_low.csv')
user_genre = pd.read_csv(r'C:\Users\nasta\OneDrive - TU Eindhoven\Desktop\DSS\user_genre.csv')
```

```python
user_tracks_high = pd.merge(user_tracks_high, user_genre, on='user_id', how='left')
user_tracks_low = pd.merge(user_tracks_low, user_genre, on='user_id', how='left')
```

Extract user tailored playlists based on the audio feature ranges and prefered genre of exploration 

```python
user_tracks_high['recommended_songs'] = np.nan

for index, row in user_tracks_high.iterrows():
    recommended_songs = sp.recommendations(seed_genres = [row['genre_name']], limit=10, 
                                           min_danceability = row['min_danceability'], max_danceability = row['max_danceability'], 
                                           min_energy = row['min_energy'], max_energy = row['max_energy'],  
                                           min_speechiness = row['min_speechiness'], max_speechiness = row['max_speechiness'], 
                                           min_acousticness = row['min_acousticness'], max_acousticness = row['max_acousticness'], 
                                           min_instrumentalness = row['min_instrumentalness'], max_instrumentalness = row['max_instrumentalness'], 
                                           min_liveness = row['min_liveness'], max_liveness = row['max_liveness'], 
                                           min_valence = row['min_valence'], max_valence = row['max_valence'])
    track_ids = []
    for track in recommended_songs['tracks']:
        track_ids.append(track['id'])
    user_tracks_high['recommended_songs'][index] = track_ids
```

```python
user_tracks_low['recommended_songs'] = np.nan
user_tracks_low['min_duration_ms'] = user_tracks_low['min_duration_ms'].astype(int)
user_tracks_low['min_key'] = user_tracks_low['min_key'].astype(int)
user_tracks_low['max_duration_ms'] = user_tracks_low['max_duration_ms'].astype(int)
user_tracks_low['max_key'] = user_tracks_low['max_key'].astype(int)

for index, row in user_tracks_low.iterrows():
    recommended_songs = sp.recommendations(seed_genres=[row['genre_name']], limit=10,
                                           min_duration_ms=row['min_duration_ms'], max_duration_ms=row['max_duration_ms'],
                                           min_key=row['min_key'], max_key=row['max_key'],
                                           min_loudness=row['min_loudness'], max_loudness=row['max_loudness'],
                                           min_tempo=row['min_tempo'], max_tempo=row['max_tempo'])
    track_ids = []
    for track in recommended_songs['tracks']:
        track_ids.append(track['id'])
    user_tracks_low['recommended_songs'][index] = track_ids
```

```python
#for each of the recommended songs list, make a playlist named 'user_id' + 'high'/'low' and add the songs to it
for index, row in user_tracks_high.iterrows():
    user_id = row['user_id']    
    
    # Skip rows with empty recommended_songs
    if not row['recommended_songs']:
        continue
    
    playlist_name = 'high_level_features_playlist' + user_id
    playlist = sp_user.user_playlist_create(username, playlist_name)
    sp_user.user_playlist_add_tracks(username, playlist['id'], row['recommended_songs']) 
```

```python
for index, row in user_tracks_low.iterrows():
    user_id = row['user_id']
    
    # Skip rows with empty recommended_songs
    if not row['recommended_songs']:
        continue
    
    playlist_name = 'low_level_features_playlist' + user_id
    playlist = sp_user.user_playlist_create(username, playlist_name)
    
    # Add tracks only if recommended_songs is not empty
    sp_user.user_playlist_add_tracks(username, playlist['id'], row['recommended_songs'])
```

```python
# Extract links of all newly created playlists starting with 'high' or 'low'
user_playlists = sp_user.user_playlists(username)
playlist_links = []

for playlist in user_playlists['items']:
    playlist_name = playlist['name'].lower()
    
    if playlist_name.startswith('high') or playlist_name.startswith('low'):
        playlist_links.append(playlist['external_urls']['spotify'])

playlist_links
```
